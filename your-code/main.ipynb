```python
# your code here
import numpy as np

def generate_uniform(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)
```

```python
# your code here
data_1 = generate_uniform(10, 15, 100)
data_2 = generate_uniform(10, 60, 1000)
```

```python
# your code here
import matplotlib.pyplot as plt

# Histograma para el primer conjunto
plt.figure()
plt.hist(data_1, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Uniform Distribution (10 to 15, count=100)")
plt.show()

# Histograma para el segundo conjunto
plt.figure()
plt.hist(data_2, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Uniform Distribution (10 to 60, count=1000)")
plt.show()
```

How are the two distributions different?

Increasing the number of samples makes the empirical distribution closer to the theoretical uniform distribution, while increasing the range spreads the data over a wider interval.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here

def generate_normal(mean, std, count):
    return norm.rvs(loc=mean, scale=std, size=count)
```

```python
# your code here
data_std_1 = generate_normal(10, 1, 1000)
```

```python
# your code here
data_std_50 = generate_normal(10, 50, 1000)
```

```python
# your code here
import matplotlib.pyplot as plt

# Histograma para std = 1
plt.figure()
plt.hist(data_std_1, bins=30)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Normal Distribution (mean=10, std=1)")
plt.show()

# Histograma para std = 50
plt.figure()
plt.hist(data_std_50, bins=30)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Normal Distribution (mean=10, std=50)")
plt.show()
```

How are the two distributions different?

Your answer here
Both distributions are centered at the same mean value of 10, so their peaks occur at the same location. However, they differ in their spread due to different standard deviations. The distribution with a standard deviation of 1 is narrow and concentrated, meaning most values are close to the mean. In contrast, the distribution with a standard deviation of 50 is much wider and more spread out, indicating high variability and a greater likelihood of extreme values. This shows that while the mean determines the center of a normal distribution, the standard deviation controls how dispersed the data is.

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. Œª = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
def generate_exponential(mean, size):
    # Generate exponentially distributed random numbers using numpy
    return np.random.exponential(scale=mean, size=size)

# Generate samples
exp_10 = generate_exponential(10, 10)
exp_100 = generate_exponential(10, 100)

# Plot histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(exp_10, bins=100, edgecolor='black')
plt.title('Exponential Distribution (mean=10, size=10)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(exp_100, bins=100, edgecolor='black')
plt.title('Exponential Distribution (mean=10, size=100)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The distribution with size=10 is more irregular and less smooth due to fewer samples, while the distribution with size=100 is smoother and better represents the exponential shape.

The mean changes, so the distribution changes as well

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
from scipy.stats import expon

# mean = 10, so lambda = 1/10
mean = 10
x = 15

# Probability that a customer spends less than 15 minutes
prob_less_than_15 = expon.cdf(x, scale=mean)
print(f"Probability that a customer spends less than 15 minutes: {prob_less_than_15:.4f}")
```
